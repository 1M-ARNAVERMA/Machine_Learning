### Linear Regression 

```python
from sklearn.linear_model import LinearRegression

x = [[1],[2],[3],[4],[5]]
y = [40,50,65,75,90]

model = LinearRegression()

model.fit(x,y)

hours = float(input("How many hours did you study ?"))

predicted_marks = model.predict([[hours]])  # 2D List

print(f"Based on your hours {hours} you may score around {predicted_marks}")
```

# Classification

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

x = [[1],[2],[3],[4],[5]]
y = [0,0,1,1,1] # Result 0 F , 1 P

model = LogisticRegression()

model.fit(x,y)

hours = float(input("How many hours did you study ?"))

result = model.predict([[hours]])[0]  # 2D List

if result == 1:
    print("You are likely to pass")
else:
    print("Your are likely to fail")
```
