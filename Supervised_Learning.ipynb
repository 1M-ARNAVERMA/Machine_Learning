### Linear Regression 

```python
from sklearn.linear_model import LinearRegression

x = [[1],[2],[3],[4],[5]]
y = [40,50,65,75,90]

model = LinearRegression()

model.fit(x,y)

hours = float(input("How many hours did you study ?"))

predicted_marks = model.predict([[hours]])  # 2D List

print(f"Based on your hours {hours} you may score around {predicted_marks}")
```
