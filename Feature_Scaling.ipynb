```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split 
```

```python
data = {
    'StudyHours':[1,2,3,4,5],
    'TestScore':[40,50,60,70,80]
}
df = pd.DataFrame(data)
```

```python
Standard_scaler = StandardScaler()   # Standard scaler, scaler object create karega
Standard_scaled = Standard_scaler.fit_transform(df)
# fit_transform - Mean and SD har columns se seekh jayega, toh iss information ka use karega to scale each value

print("Standard Scaler Output : ")
print(pd.DataFrame(Standard_scaled, columns=['StudyHours','TestScore']))
```

```python
minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df)

print('\n MinMax Scaled Output')
print(pd.DataFrame(minmax_scaled, columns=['StudyHours','TestScore']))
```

```python
X = df[['StudyHours']]
Y = df[['TestScore']]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
# Random state 42 ka matlab hai ki same result lana hai 

print('Training Data')
print(X_train)

print('Test Data')
print(X_test)
```
