```python
import pandas as pd
```

```python
data = {
    'Name':['Pavan','kapil','Lalit','Ishan','Om'],
    'Age' : [25,None,44,23,None],
    'Salary' : [50000, 60000 , 70000 , None , None]
}

df = pd.DataFrame(data)
print("Original Dataframe")
print(df)
```

```python
print(df.isnull().sum())
# Har column mai kitne missing values hai 
```

```python
df.isnull().mean() * 100
# This will tell us how much data ( % ) is missing currently
```

```python
df_drop = df.dropna()
print(df_drop)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace = True)
# Age ke empty entiries mai, Age ka mean value place kardo
df['Salary'].fillna(df['Salary'].mean(), inplace = True)
# Salary ke empty entiries mai, Salary ka mean value place kardo
print(df)
```

```python
df.isnull().mean() * 100
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

df= pd.read_csv('student_dataset.csv') 
```

```python
df_label = df.copy()  # Safe duplicate data bana rahe hai for our original data

le = LabelEncoder()
df_label['Gender_Encoded'] = le.fit_transform(df_label['Gender'])
# Fit - Jo values hai unhe seekhna hai 
# Transform - Seekhne ke baad number se replace kar dena hai
df_label['Passed_Encoded'] = le.fit_transform(df_label['Passed'])
print('\n Label Encoded Data')
print(df_label[['Name','Gender','Gender_Encoded','Passed','Passed_Encoded']])
```

```python
df_encoded = pd.get_dummies(df, columns = ['City']) # Dummies - Text wale columnn ko small binary number mai convert kar do
print('\n One-Hot Encoded Data')
print(df_encoded)
```
