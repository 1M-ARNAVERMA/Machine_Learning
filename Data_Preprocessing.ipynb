```python
import pandas as pd
```

```python
data = {
    'Name':['Pavan','kapil','Lalit','Ishan','Om'],
    'Age' : [25,None,44,23,None],
    'Salary' : [50000, 60000 , 70000 , None , None]
}

df = pd.DataFrame(data)
print("Original Dataframe")
print(df)
```

```python
print(df.isnull().sum())
# Har column mai kitne missing values hai 
```

```python
df.isnull().mean() * 100
# This will tell us how much data ( % ) is missing currently
```

```python
df_drop = df.dropna()
print(df_drop)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace = True)
# Age ke empty entiries mai, Age ka mean value place kardo
df['Salary'].fillna(df['Salary'].mean(), inplace = True)
# Salary ke empty entiries mai, Salary ka mean value place kardo
print(df)
```

```python
df.isnull().mean() * 100
```
