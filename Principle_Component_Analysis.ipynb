```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler  # SS is used to scale data -> Same scale ya range follow kare
from sklearn.decomposition import PCA # Decomp -> Bade data ko chote components mai break kar dega
```

```python
data ={
    'Age' : [25,30,35,40,45,50],
    'Income' : [30000,40000,50000,60000,70000,80000],
    'Spending' : [70,60,50,40,30,20],
    'Savings' : [1000,5000,8000,10000,15000,20000]
}

df = pd.DataFrame(data)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
pca = PCA(n_components = 2)
pca_result = pca.fit_transform(scaled_data)
```

```python
pca_df = pd.DataFrame(pca_result, columns = ['PCA1','PCA2'])
```

```python
explained_variance = pca.explained_variance_ratio_ 
# Kitna % of total information, every component has captured
```

```python
print("Variance Captured by each PCA Component: ")
print(np.round(explained_variance * 100,2))
```

```python
plt.figure(figsize = (8,6))
plt.scatter(pca_df['PCA1'],pca_df['PCA2'],color = 'black')
plt.title("PCA projection (2D View)")
plt.xlabel('PCA1 main pattern')
plt.ylabel('PCA2 minor pattern')
plt.grid(True)
plt.show()

print("New Data with 2 features")
print(pca_df)
```
